```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

data = pd.read_csv("internship_train.csv")
data.describe().transpose()
```

```python
#Ciekawie wyglda jedynie parametr 6., caa reszta wydaje si by szumem o rnych wartociach. 
#Na wszelki wypadek jeszcze rzut oka na rozkady interesujcych zmiennych
a = abs(data.corr())
l = 53
for i in range(l+1):
    for ii in range(l+1):
        if (((i!=ii) and a.iloc[i,ii]>0.011 )):
            print(i, ii, a.iloc[i,ii])
            plt.scatter(data=data, x=data.iloc[:,i], y=data.iloc[:,ii])
            plt.show()
```

```python
#Bardziej szczegowy rzut oka
ciekawe = [6, 8, l]
for c, i in enumerate(ciekawe[0:-1]):
    for ii in ciekawe[c:]:
        if i<ii:
            print(i, ii)
            plt.scatter(data=data, x=data.iloc[:,i], y=data.iloc[:,ii])
            plt.show()
            if ii == 53:
                plt.hist(x=data.iloc[:,i])
                plt.show()
```

```python
#Po analizie powyszych wykresw do predykcji zmiennej target proponuj model kwadratowy oparty o zmienn 6
#Na uwag zasuguje te zmienna 8, ktra wydaje si odpowiada "pozytywnoci" zmiennej 6
x, y = data[[ str(i) for i in range(l-1)]], data['target']
plt.scatter(x['6'], y)
model6 = np.polyfit(x['6'], y, 2)
x_range = range(-10, 11)
predict = np.poly1d(model6)
plt.plot(x_range, predict(x_range), c = 'r')
```

```python
#Jak widzimy model si idealnie wpasowuje w treningowe dane. Policzmy rmse
def rmse(a, b):
    return(np.sqrt(np.mean(a - b) ** 2))
print(rmse(predict(x['6']), y))
```

```python
#Jednak dla pewnoci podzielmy zbir na uczcy i testowy:
msk = np.random.rand(len(data)) < 0.8
train = data[msk]
val = data[~msk]

x_train, y_train = train['6'], train['target']
x_test, y_test = val['6'], val['target']

model = np.polyfit(x_train, y_train, 2)
predict = np.poly1d(model)

plt.scatter(x_test, y_test)
plt.plot(x_range, predict(x_range), c='r')
plt.show()

print(rmse(predict(x_train),y_train))
print(rmse(predict(x_test), y_test))
print(rmse(predict(data['6']), data['target']))
```
